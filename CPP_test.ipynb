```python
from CPP_makeplot import *
mPlots, vVar = make_plot(iX=50)
mGPlots = cal_gamma(mPlots)
```

```python
# Make plot of ruin probabilities
vUarray = get_Uarray()
for iIndex, vPlot in enumerate(mPlots):
    plt.plot(vVar,vPlot, label = f'$u = {vUarray[iIndex]}$')
plt.xlabel(r"Var$(\Omega_{app})$")
plt.ylabel(r"$p(u|\Omega_{app})$")
plt.legend(loc="upper right")
plt.yscale("log") 
plt.tight_layout()
plt.savefig('CPP-prob.pdf') 

```

```python
# Make plot simulated gamma
vUarray = get_Uarray()
for iIndex, vGPlot in enumerate(mGPlots):
    plt.plot(vVar,vGPlot, label = f'$u = {vUarray[iIndex]}$')
plt.xlabel(r"Var$(\Omega_{app})$")
plt.ylabel(r"$p(u | \Omega_{app})e^{\theta^* u}$")
plt.legend(loc="upper right")
plt.tight_layout()
plt.savefig('CPP-gamma.pdf') 
```
