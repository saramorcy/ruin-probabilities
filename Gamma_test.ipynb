## The Code below checks the sensitivity for floating point errors for the derivatives of $\xi_{Q,k}(\theta + \epsilon, \omega)$

```python

def D_Zeta_Q_k_sensitivity(parameters,iK):
    """
    For: Hyp-Erlang distributed inter inspection times.
    This function returns the zeroth up to the kth derivative in a vector of xi_Q,k evaluated at theta.
    """
    alpha = sp.Symbol('alpha')
    fW = parameters[0]
    iK = int(iK)
    fTheta = Cal_Theta(parameters)
    fBeta = fW
    fPsi_Beta_q = Psi_q_S(fBeta)
    Teller_1 = (alpha - fPsi_Beta_q)*fBeta
    Noemer_1 = (phi_q(alpha) - fBeta)*(fPsi_Beta_q)
    Teller = sp.Pow(Teller_1, iK)
    Noemer = sp.Pow(Noemer_1, iK)
    fVal_Zeta_Q_k = Teller/Noemer    
    
    # Make list of N and T values
    lN = [Noemer]
    lT = [Teller]
    for n in range(1,iK+1):
        lN.append(sp.diff(lN[-1], alpha))
        lT.append(sp.diff(lT[-1], alpha))
        
    # Make list of Zeta values
    lDZeta = [Teller/Noemer, (lT[1] - lN[1]*fVal_Zeta_Q_k)/lN[0]]
    for n in range(2,iK+1):
        som = 0
        for i in range(1,n+1):
            som += comb(n,i)*lN[i]*lDZeta[n-i]
        Simplified_DZeta = (lT[n] - som)/lN[0]
        lDZeta.append(Simplified_DZeta)
    
    # eps_steps = np.linspace(float(fTheta)-0.1,float(fTheta)+0.1,20)
    eps_steps = np.linspace(-0.06,0.06,20)
    lDZeta_Theta = []
    for itemDZeta in lDZeta:
        lDZeta_Theta_eps = []
        for eps in eps_steps:
            plus = itemDZeta.subs(alpha, fTheta + eps)
            lDZeta_Theta_eps.append(plus)
        lDZeta_Theta.append(lDZeta_Theta_eps)
    return eps_steps, lDZeta_Theta

# parameters = search_par_hypErlang(0.16393878)
parameters = search_par_hypErlang(0.17)
print(f"k = {parameters[1]}, \omega = {parameters[0]}, p = {parameters[2]}")
eps_steps, plot = D_Zeta_Q_k_sensitivity(parameters,parameters[1])
# Make plot for sensitivity
for iIndex, Der in enumerate(plot):
    plt.plot(eps_steps,Der, label = f'm = {iIndex}')
plt.xlabel(r"$\epsilon$")
plt.ylabel(r"$\xi_{Q,6}^{(m)}(\theta + \epsilon,\omega)$")
plt.legend(loc="upper right") 
plt.title(r"Derivatives of $\xi_{Q,6}(\theta + \epsilon,\omega)$ where $\omega \approx 5.94$ ")
plt.savefig('sensitivity-k6.pdf') 
plt.tight_layout()
```
